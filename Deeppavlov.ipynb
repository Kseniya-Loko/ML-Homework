```python
import nltk
text_arr=[]


tokArray =[]
ner_array=[]
nltk.download('punkt')
import string
import nltk
from nltk.tokenize.toktok import ToktokTokenizer
toktok = ToktokTokenizer()

from tqdm import tqdm
translator = str.maketrans('', '', string.punctuation) #убираем всю лишнюю пунктуацию из текста

with tqdm(total = len(med)) as pbar: #преобразуем наш прочитаный как df файл в токенизированные предложения

        for index,elem in med.iterrows():
            pbar.update(1)

            sent_text = nltk.sent_tokenize(str(elem['text'])) # this gives us a list of sentences
            for senten in sent_text:
                tokArray.append(tuple((senten,elem['text']))) #получаем список из них tokArray
```

```python
from concurrent.futures import ThreadPoolExecutor as PoolExecutor
import http.client
import socket
toker = [] #массив, куда мы сохраняем результат
def get_it(url): #функция для распараллеливания процессов
        try:
            toker.append(tuple(((ner_model([url[0]][:500]),url[1])))) #url0 - разбитое предложение, url1 - исходный текст
        except RuntimeError:
            print('auch')

with PoolExecutor(max_workers=16) as executor:
    for _ in executor.map(get_it, tokArray): #обработка текста - 1 арг. функция, 2 аргумент - токенизированные предложения
        pass
```

```python
def nercenthandler (nerdataname):
    my_arr=[]
    with tqdm(total = len( nerdataname)) as pbar:
        indexer=0
        for elem in nerdataname:
            pbar.update(1)

            indexer+=1
            locindex=0
            person=''
            stringer= ' '.join(elem[0][0][0])
                
                
            #sent = sent_model([stringer])[0][0]
            if len(elem[0][1]) > 0: 
                for inder, val in enumerate(elem[0][1][0]):
                    for tag in ['PERSON','ORG','DATE','GPE','NORP','QUANTITY','WORK_OF_ART','CARDINAL','MONEY','ORDINAL','TIME','FAC','LAW','PRODUCT','LANGUAGE','EVENT','LOC']:
                        if 'B-'+tag ==val:
                                tagvalue = elem[0][0][0][inder]
                                try:
                                    if (elem[0][1][0][inder+1]!='I-'+tag):
                                        my_arr.append(tuple((tagvalue,elem[1],tag.lower(),str(indexer))))
                                except IndexError:
                                    #my_arr.append(tuple((tagvalue,elem[1],elem[2],tag.lower(),cntspeech,str(indexer)+str(locindex))))
                                    pass
        
        
                        elif 'I-'+tag == val:
                                if elem[0][0][0][inder][0] in string.punctuation or elem[0][0][0][inder-1][-1] in string.punctuation:
                                    tagvalue  = tagvalue + ''+elem[0][0][0][inder]
                                else:
                                    tagvalue  = tagvalue + ' '+elem[0][0][0][inder]
                                try:
                                    if  elem[0][1][0][inder+1] !='I-'+tag:
                                        my_arr.append(tuple((tagvalue,elem[1],tag.lower(),str(indexer))))
                                except IndexError: 
                                    pass
    return my_arr
```

```python
#вопрос-ответная модель, где squad - датасет, на котором она обучена

from deeppavlov import build_model, configs

model = build_model(configs.squad.squad_torch_bert, download=True)
```

```python
from tqdm import tqdm

nerframeper['relation']=''
with tqdm(total = len(nerframeper)) as pbar:

    for index, row in nerframeper.iterrows():
        pbar.update(1)
        nerframeper.at[index,'relation']=model([row['sent']], ['Кто'+row['nervalue']+'?'])[0][0]
```
